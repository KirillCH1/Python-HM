```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})

result = []

for i in lst:
  if i == 'robot':
    result.append(["0" , " 1"])
  if i == 'human':
    result.append(["1" , " 0"])

data = pd.DataFrame(result, columns=['human' , 'robot'] )
data.head()
```
